#**Análise da Evolução do Rendimento Médio Habitual da População Ocupada no Brasil**

**Objetivo: avaliar a evolução do rendimento médio habitual da população ocupada em relação aos diferentes setores.**

A hipótese de trabalho é verificar a tendência de comportamento dos rendimentos médios e estabeler relação com fatores que possam ter contribuido para o aumento ou redução do indicador no seu respectivo segmento.

# **1. Problema de pesquisa**

Inicialmente, é necessário esclarecer os conceitos utilizados na estruturação dos dados obtidos:

**População ocupada**

---


São classificadas como ocupadas na semana de referência as pessoas que, nesse
período, trabalharam pelo menos uma hora completa em trabalho remunerado em dinheiro, produtos, mercadorias ou benefícios (moradia, alimentação, roupas, treinamento etc.) ou em trabalho sem remuneração direta em ajuda à atividade econômica de membro do domicílio ou, ainda, as pessoas que tinham trabalho remunerado do qual estavam temporariamente afastadas nessa semana.

Consideram-se como ocupadas temporariamente afastadas de trabalho remunerado as pessoas que não trabalharam durante pelo menos uma hora completa na semana de
referência por motivo de: férias, folga, jornada de trabalho variável, licença maternidade e fatores ocasionais.

Assim, também foram consideradas as pessoas que, na data de referência, estavam, por período inferior a 4 meses: afastadas do trabalho em licença remunerada por motivo de doença ou acidente da própria pessoa ou outro tipo de licença remunerada; afastadas do próprio empreendimento sem serem remuneradas por instituto de previdência; em greve ou paralisação. Além disso, também, foram consideradas ocupadas as pessoas afastadas por motivos diferentes dos já citados, desde que tivessem continuado a receber ao menos uma parte do pagamento e o período transcorrido do afastamento fosse inferior a 4 meses.

**Rendimento do Trabalho**

---
Rendimentos habituais: O rendimento habitual consiste no rendimento recebido por
empregados, empregadores e trabalhadores por conta própria, mensalmente, sem acréscimos extraordinários ou descontos esporádicos.

Para o empregado, o rendimento mensal habitualmente recebido exclui todas as parcelas que não tenham caráter contínuo (bonificação anual, salário atrasado, horas extras, participação anual nos lucros, 13o salário, 14o salário, adiantamento de salário, etc.) e não considera os descontos ocasionais (faltas, parte do 13o salário antecipado, prejuízo eventual causado ao empreendimento etc.).

Caso o rendimento recebido de empregado, de trabalhador por conta própria e de
empregador seja variável, considera-se o rendimento habitual aquele em média recebido pela pessoa no período em que realizava o trabalho declarado na semana de referência. Quando a remuneração varia em função do período ou estação do ano, considera-se o rendimento mensal que a pessoa ganha habitualmente nesse período sazonal.

**Diante disso, qual será a trajetória do rendimento médio nos diferentes setores considerando o histórico das valores percebidos e o panorama econômico?**

## **1.1 Objetivos gerais**

Avaliar o comportamento dos rendimentos médios percebidos pela população ocupada no Brasil.

## **1.2 Objetivos específicos**

Estabelecer uma análise preditiva na evolução dos rendimentos médios. Caso sejam verificadas variações significativas, verificar o contexto econômico do país no período.

# **2. Importância do problema**

As variações desse indicador estão vinculados às condições de trabalho, desigualdades de renda e oferta de emprego. Nos níveis neste estudo, as diferentes atividades econômicas refletem as remunerações percebidas de acordo com suas inerentes características.

#**3. Hipótese**

A hipótese central é a interdependência dos setores de acordo com o ambiente econômico e o impacto nas remunerações que serão refletidas no rendimento médio da população ocupada.

**5.9 Autocorrelação dos dados**

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.api as sm
from matplotlib.collections import PolyCollection, LineCollection

first_diff = df.shift(1)
first_diff = first_diff.dropna(inplace = False)

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,6), dpi= 80)
plot_acf(first_diff['24387 Conta propria'].values.squeeze(), lags=30, ax=ax1,title= "Autocorrelação - Conta Própria", alpha=0.05,vlines_kwargs=dict(linewidth=.5))
my_color="red"
for item in ax1.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax1.lines:
  item.set_color(my_color)

plot_pacf(first_diff['24387 Conta propria'].values.squeeze(), lags=30,ax=ax2,method='ywmle', title= "Autocorrelação parcial - Conta Própria",alpha=0.05)
my_color="red"
for item in ax2.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax2.lines:
  item.set_color(my_color)

plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,6), dpi= 80)
plot_acf(first_diff['24388 Empregadores'].values.squeeze(), lags=30, ax=ax1,title= "Autocorrelação - Empregadores", alpha=0.05,vlines_kwargs=dict(linewidth=.5))
my_color="yellow"
for item in ax1.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax1.lines:
  item.set_color(my_color)

plot_pacf(first_diff['24388 Empregadores'].values.squeeze(), lags=30,ax=ax2,method='ywmle', title= "Autocorrelação parcial - Empregadores",alpha=0.05)
my_color="yellow"
for item in ax2.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax2.lines:
  item.set_color(my_color)

plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,6), dpi= 80)
plot_acf(first_diff['24385 Setor privado'].values.squeeze(), lags=30, ax=ax1,title= "Autocorrelação - Setor Privado", alpha=0.05,vlines_kwargs=dict(linewidth=.5))
my_color="blue"
for item in ax1.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax1.lines:
  item.set_color(my_color)

plot_pacf(first_diff['24385 Setor privado'].values.squeeze(), lags=30,ax=ax2,method='ywmle', title= "Autocorrelação parcial - Setor Privado",alpha=0.05)
my_color="blue"
for item in ax2.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax2.lines:
  item.set_color(my_color)

plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,6), dpi= 80)
plot_acf(first_diff['24386 Setor publico'].values.squeeze(), lags=30, ax=ax1,title= "Autocorrelação - Setor Publico", alpha=0.05,vlines_kwargs=dict(linewidth=.5))
my_color="green"
for item in ax1.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax1.lines:
  item.set_color(my_color)

plot_pacf(first_diff['24386 Setor publico'].values.squeeze(), lags=30,ax=ax2,method='ywmle', title= "Autocorrelação parcial - Setor Público",alpha=0.05)
my_color="green"
for item in ax2.collections:
  if type(item)==PolyCollection:
    item.set_facecolor(my_color)
  if type(item)==LineCollection:
    item.set_color(my_color)
for item in ax2.lines:
  item.set_color(my_color)

plt.show()
```
